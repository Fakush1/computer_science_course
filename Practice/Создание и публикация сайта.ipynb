## HTML элементы

### Теги

В программно коде задания вы видели символы HTML, заключенные в угловые скобки - `<html>`, `<body>`. Это так называемые `теги` языка разметки.

Элементы HTML обычно состояит из двух тегов, одного открывающегося и одного закрывающегося. В нашем примере `<html>` - открывающийся тег, а `</html>` - закрывающийся тег.

Текст между открывающимся и закрывающимся тегами является содержимым HTML-документа. С помощью `<html>` и `</html>` браузеру сообщается о начале и конце HTML разметки на веб-странице.

### Атрибуты

`Атрибуты` объявляют дополнительную информацию об элементе HTML. Атрибуты являются частью открывающегося тега элемента и всегда имеют `имя` и `значение`.

В качестве примера рассмотрим элемент HTML для ссылки.

```<a href = "https://github.com/SerjiEvg/computer_science_course"> Репозиторий курса </a>```

Приведенный выше элемент `<a>` включает `href` атрибут (сокращение от hyperlink) со значением https://github.com/SerjiEvg/computer_science_course. Браузер знает, что он должен отображать ссылку как <a href = "https://github.com/SerjiEvg/computer_science_course"> Репозиторий курса </a>. 

### Базовая структура HTML-страницы

Мы уже использовали теги `<html>` и `<body>`. Но базовая структура HTML-страницы обычно содержит еще несколько. 

Рассмотрим пример кода.



````python
<! DOCTYPE html> 
<html> 
    <head> 
        <meta charset = "utf-8" > 
        <title> Веб-портфолио </title> 
    </head> 
    <body> 
        <h1> Веб-портфолио </h1> 
        <p> Напишите все, что хотите рассказать миру. </p> 
    </body> 
</html>
````

Несколько рекоммендаций и пояснений по оформлению этого небольшого кода:
- Всегда ставьте тег `<!DOCTYPE html>` в первую строчку. Он сообщает браузеру о типе документа.

- Тег `<html>` указывает на начало, а `</html>` - на конец документа.

- Элемент `<head>` содержит дополнительную информацию о странице. В отличие от элемента `<body>`, эта информация не отображается в основной области браузера.

- Внутри `<head>`, должно быть указание о наборе символов: `<meta charset="utf-8">`. Если вы не укажете набор символов, некоторые специальные символы могут отображаться неправильно. Вы могли заметить, что у `<meta>`элемента нет закрывающего тега. Есть несколько элементов без закрывающих тегов (`<br>`, `<img>`и т.д.), Они являются исключением из общего правила.

- Далее вы увидите `<title>`элемент. Заголовок отображается в строке заголовка в верхней части окна браузера.

- Все, что `<body>`находится внутри элемента, отображается в основной области браузера.

- А `<h1>` определяет основной заголовок. Суб-заголовки могут быть созданы с помощью тегов `<h2>`, `<h3>`, `<h4>`, `<h5>`, и `<h6>`.

- Текст между `<p>` и `</p>` является абзацем.

- После каждого открывающего тега следующий элемент должен иметь отступ (с табуляцией или двумя пробелами) для лучшего чтения кода. Убедитесь, что вы следуете этой привычке.

### Вставка изображения

Чтобы добавить изображение, мы используем элемент `<img>`. В следующем примере добавляется изображение аватарки

```<img src = "https://github.com/SerjiEvg/computer_science_course/blob/master/Assigments%20for%20practice/images/avatar.png?raw=true" alt = "Изображение аватарки">```

У элемента `<img>` только один открывающийся тег, но нет закрывающегося. Он содержит атрибуты `src` и `alt`.

Атрибут `src` указывает URL, то есть местоположение и имя файла изображения.

Атрибут `alt` является альтернативным текстом, который описывает содержимое изображения (нужен в том случае, если изображение не будет подгружено).

### Относительные и абсолютные URL-адреса

## Подготовка сайта к публикации

URL-адреса используются для атрибута `src` изображения, а также для атрибута `href` ссылок.

URL-адрес указывает местоположение файла и в зависимости от его близости может указываться `относительный` или `абсолютный` URL.

Если файл является частью того же веб-сайта, можно использовать относительный URL адрес. Относительный URL указывается по отношению либо к текущей HTML странице или по отношению к корню сайта. 

Например, можно указать относительный адрес для загрузки аватарки ```<img src = "../Assigments%20for%20practice/images/avatar.png?raw=true" alt = "Изображение аватарки">```

Указание `../` возвращает нас к родительскому каталогу (на уровень вверх). Если путь нужно начать с текущего каталога и двигаться во вложенные папки, то просто указывается `/`.

<img src = "../Assigments%20for%20practice/images/avatar.png?raw=true" alt = "Изображение аватарки">

Если файл находится на другом веб-сайте, то необходимо использовать абсолютный URL-адрес.

## Публикация сайта

### Хостинг на Netlify

Что нужно, чтобы наш сайт можно было просматривать в Интернете?

Нам нужен веб-сервер, на котором хранится наш веб-сайт, и адрес (URL), чтобы наш веб-сайт можно было найти.

Есть множество сервисов, где можно арендовать серверное пространство. Такая услуга называется `веб-хостингом`. Мы рассмотрим несколько вариантов бесплатного хостинга.

Наиболее удобный для нас вариант - [Netfly](https://www.netlify.com/). Это простой сервис для быстрого развертывания и публикации сайтов.

Чтобы опубликовать наш сайт нужно:
- Зарегистрироваться на [сайте Netlify](https://www.netlify.com/). Перейдите по ссылке [Sign Up](https://app.netlify.com/signup?_ga=2.234917896.861481127.1634358840-58973899.1634066246) на главной странице сервиса.

- Можете войти через Github, если у вас уже имеется аккаунт. Или выберите вход через Email (нужно будет указать свой email и задать пароль).
